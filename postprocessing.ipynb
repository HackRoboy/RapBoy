```python
import librosa
import os
import numpy as np
import soundfile as sf

from tts import TTS, Voice
```

```python
tts = TTS(Voice(Voice.Language.enUS, Voice.Sex.male, "Justin"))
text = "Hello there. I am Roboy"
# samplerate - number of discrete points in second for the sound wave, Hz
y, sr = sf.read(tts.speak(text))
```

```python
from IPython.display import Audio
Audio(data=y, rate=sr)
```

```python
# ending part
size_to_modify = int(len(y)*0.30)
end_range_start = len(y) - size_to_modify
end_ind = list(range(end_range_start, len(y)))
y_ending = y[end_ind]
y_ending.shape
```

```python
# mid part
mid = int(end_range_start/2)
mid_range_start = mid - int(size_to_modify/2)
mid_range_end = mid + int(size_to_modify/2)
mid_ind = list(range(mid_range_start, mid_range_end))
y_mid = y[mid_ind]
y_mid.shape
```

```python
y_end_shifted = librosa.effects.pitch_shift(y_ending, sr, n_steps=-1)
y_end_shift_stretch = librosa.effects.time_stretch(y_end_shifted, 0.5)
```

```python
y_mid_shifted = librosa.effects.pitch_shift(y_mid, sr, n_steps=-1)
y_mid_shift_stretch = librosa.effects.time_stretch(y_mid_shifted, 0.5)
```

```python
y_cut = y[:end_range_start]
print(y_cut.shape)
y_firsth = y_cut[:mid_range_start]
y_secondh = y_cut[mid_range_end:]
print(y_firsth.shape)
print(y_secondh.shape)
```

```python
y_modified = np.concatenate([y_firsth, y_mid_shift_stretch, y_secondh, y_end_shift_stretch])
```

```python
librosa.output.write_wav('modified.wav', y_modified, sr)
```

```python
os.system('mpv' + ' ' + 'modified.wav')
```
